```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/50_Startups (1).csv")
dataset
```

```python
dataset.columns
```

```python
d = pd.get_dummies(dataset, columns=['State'], dtype=int)
d
```

```python
d.columns
```

```python
x=d[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
y=d[['Profit']]

# Train,Test,Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.3, random_state = 45)

# Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# Weight and Bias
Weight = model.coef_
w = Weight
print("The Weight of the model is ",w)

Bias = model.intercept_
b = float(Bias)
print("\nThe Bias of the model is ",b)

# Evaluation Metrics
from sklearn.metrics import r2_score
y_p = model.predict(x_test)
r_score = r2_score(y_test,y_p)
print("\nThe accuracy of the model is ",r_score)
print()
```
